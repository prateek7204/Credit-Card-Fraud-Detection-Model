# CREDIT CARD FRAUD DETECTION USING LOGISTIC REGRESSION

```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"C:\Users\HP\Desktop\Data related files\Credit Card Fraud Dataset\creditcard.csv")
df.head()
```

```python
real = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
real.shape
```

```python
fraud.shape
```

```python
real.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
#as the data is imbalanced as we can see that the real cases are over 2 lakhs and the fraudulent cases are only 492 so we are
#going to take any random 492 

real_sample = real.sample(n=492) #random sampling
```

```python
#concatinating two datasets 
new = pd.concat([real_sample, fraud], axis = 0) #values will be added row wise
```

```python
new.head()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

```python
X = new.drop(columns = 'Class', axis = 1)
Y = new['Class']
```

```python
X.head()
```

```python
Y.tail()
```

```python
#splitting the data into training and testing 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2 , stratify = Y, random_state = 2)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
#model training 
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('The Accuracy of the Training Model is : ', training_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('The Accuracy of the Testing Model is : ', testing_accuracy)
```
